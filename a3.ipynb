```python
def partitionIP(lst, l, r):
    pivot = lst[l]
    i = l
    j = r + 1
    while True:
        while True:
            i = i + 1
            if lst[i] > pivot or i >= len(lst) - 1:
                break
        while True:
            j = j - 1
            if lst[j] <= pivot:
                break
        if i < j:
            lst[i], lst[j] = lst[j], lst[i]
        else:
            lst[l], lst[j] = lst[j], lst[l]
            return j


def quicksortIP(lst: list, l: int, r: int):
    if r - l > 1:
        pivot = partitionIP(lst, l, r)
        quicksortIP(lst, l, pivot - 1)
        quicksortIP(lst, pivot + 1, r)
    else:
        return

lst = [8, 5, 1, 3, 2]
quicksortIP(lst, 0, len(lst) - 1)
lst
```
