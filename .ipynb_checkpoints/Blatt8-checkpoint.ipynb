```python
import graphviz as gr
import itertools

class Trie: 
    def __init__(self,key = "0"):
        self.key = key 
        self.val = None
        self.children = {} 
        
    def insert(self, key, val): 
        if not key: # an der stelle angekommen:
            self.val = val
        else: 
            if key[0] not in self.children:
                self.children[key[0]] = Trie(key[0])
            self.children[key[0]].insert(key[1:], val)
        
    def search(self, key):
        if len(key)==0: # Stelle gefunden
            return self.val
        elif key[0] in self.children: # wenn ich mit dem nächsten key-Zeichen weiterkomme:
            return self.children[key[0]].search(key[1:]) # rekursiver Aufruf von search auf den Kind-Trie
        else: # Es gibt keinen solchen Schlüssel
            return None # oder KeyError-Exception

    def searchIt(self,key):
        t = self
        while True:
            if len(key) == 0:
                return t.val
            elif key[0] not in t.children:
                return None
            else:
                t = t.children[key[0]]
                key = key[1:]

    def delete(self,key):
        p = self
        c = None
        t = self
        if len(key) != 0:
            c = key[0]
        while True:
            if len(key) == 0:
                if len(t.children) == 0:
                    break
                t.val = None
                return
            elif len(t.children) > 1 or t.val:
                p = t
                c = key[0]
            t = t.children[key[0]]
            key = key[1:]

        p.children.__delitem__(c)

    def keysT(self,sAcc):
        if self.val == None: s = []
        else: s = [sAcc]
        l = [self.children[c].keysT(sAcc + c) for c in self.children]
        return s + list(itertools.chain(*l))

    def keys(self):
        return self.keysT("")

    def plotPrep(self):
        edges = []
        nodes = []
        if len(self.children) == 0:
            return edges,[self.key]
        else:
            for i in self.children:
                edges.append((self.key,self.children[i].key))
                nodes.append(self.key)
                edgesR, nodesR = self.children[i].plotPrep()
                nodes += nodesR
                edges += edgesR
            return edges,nodes

    def plot(self):
        g = gr.Digraph()
        e,n = self.plotPrep()
        c = 0
        for i in e:
            g.edge(i[0],i[1],i[1])
            c += 1
        return g
```

```python
t = Trie()
t.insert("gans", 2)
t.insert("ganz", 2)
t.insert("galle", 2)
t.insert("leber", 2)
t.insert("lesen", 2)
t.insert("lesezeichen", 2)
t.insert("zeichnen", 2)
t.insert("zeilenweise", 2)
t.insert("adam", 2)
t.insert("aaron", 2)
t.keys()
```
