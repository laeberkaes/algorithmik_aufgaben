```python
Infty = float("inf")

from random import random

def randHeight(p = 0.5): # Geometrische Verteilung mit Parameter p = 1/2
    i = 1
    while random()<=p: 
        i = i+1
    return i
```

```python
class SLElement: 
    def __init__(self, key, ptrs=[], val=None): 
        self.key = key
        self.ptrs = ptrs
        self.val = val
    def __repr__(self): 
        return "(" + str(self.key) + ((", " + str(self.val)) if self.val != None else "") + "|" + str(len(self.ptrs)) + ")"

class SkipList: 
    def __init__(self, MAXHEIGHT=20): 
        self.MAXHEIGHT = MAXHEIGHT
        self.tail = SLElement(Infty)
        self.head = SLElement(None, [self.tail for _ in range(self.MAXHEIGHT)])
        self.height = 0 # maximaler Index des höhsten Elements

    def __str__(self):
        el = self.head
        erg = "[ "
        while el.ptrs[0].key < Infty:
            el = el.ptrs[0]
            erg += str(el) + ", "
        erg = erg[:-2] + " ]"
        return erg

    def keys(self):
        k_list = list()
        el = self.head.ptrs[0]
        while el.key != Infty:
            k_list.append(el.key)
            el = el.ptrs[0]
        return k_list

    def vals(self):
        v_list = list()
        el = self.head.ptrs[0]
        while el.key != Infty:
            v_list.append(el.val)
            el = el.ptrs[0]
        return v_list
    
    def insert(self,k,v = None): 
        updatePtrs = [self.head]*self.MAXHEIGHT
        el = self.head
        for ebene in range(self.height-1, -1, -1): 
            while el.ptrs[ebene].key < k: 
                el = el.ptrs[ebene]
            updatePtrs[ebene] = el
        el = el.ptrs[0]
        if el.key == k: 
            el.val = v
        else: 
            # neues Element einfügen
            newHeight = min(randHeight(), self.MAXHEIGHT)
            self.height = max(self.height, newHeight)
            # Jetzt erzeugen wir das einzufügende Element
            newEl = SLElement(k, [updatePtrs[ebene].ptrs[ebene] for ebene in range(newHeight)], v)
            # Jetzt noch die Zeiger der bisherigen Elemente umbiegen
            for ebene in range(newHeight): 
                updatePtrs[ebene].ptrs[ebene] = newEl
    
    def search(self, k): 
        el = self.head # Das Element, wo sich meine Suche gerade befindet.
        for ebene in range(self.height-1, -1,-1):
            while el.ptrs[ebene].key < k: 
                el = el.ptrs[ebene]
        el = el.ptrs[0]
        if el.key == k: 
            return el.val
        # else: Nicht gefunden, liefere nichts zurück.
```

```python
from numpy.random import choice
rnd = choice(list(range(100)),100)
```

```python
sl = SkipList()

for i in rnd: sl.insert(i)
```

```python
print(sl)
```

```python
# %timeit sl.search(961)
```

```python
def search(lst,k):
    for i in lst:
        if i == k:
            return k
```
