```python
def sum(n):
    if n == 1:
        return 1
    return n + sum(n-1)

sum(5)
```

## b)

```python
def lenR(lst):
    if len(lst) == 1:
        return 1
    return lenR(lst[:-1]) + 1

lenR([1,2,3,4])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

def line(x,h): 
    """ zeichne Linie an der Stelle x mit der Hoehe h"""
    plt.plot([x,x],[0,h], "b-")

def lineal(x,y,h):
    if h < 1: 
        return
    # auch möglich: if y-x < eps: return
    else: 
        m = (x+y)/2
        line(m,h)
        lineal(x,m, h-1)
        lineal(m,y, h-1)

#lineal(0,100,8) ## 2^8 - 1 Striche

from numpy import arange

def lineal_iter(x,y,h):
    for i in arange(0,h):
        for j in arange(x + ((y-x)/(2**h)),y,(y-x)/(2**h)):
            if (j-x) % ((y-x)/(2**(i+1))) == 0:
                line(j,h-i)

plt.figure(figsize=(20,2))                
lineal_iter(-10,100,8)
```

```python
def linealIt(x,y,h):
    listOfMid = []
    listOfTemp = []
    listOfMid.append((x + y) / 2)
    listCounter = 0
    for height in range(h,0,-1):
        for middle in listOfMid:
            line(middle,height)
            listOfTemp.append((x + middle) / 2)
            listOfTemp.append((middle + y) / 2)
            listCounter += 1
        listOfMid = listOfMid[listCounter:] + listOfTemp
        listCounter = 0
    listOfTemp = []

linealIt(0,100,8)
```

# 2.3

```python
from graphics import *

win = GraphWin(width = 500, height = 500) # create a window
win.setCoords(0, 0, 10, 10) # set the coordinates of the window; bottom left is (0, 0) and top right is (10, 10)
def drawRectangle(m:tuple, w:int):
    Rectangle(Point(m[0]-(w/2),m[1]-(w/2)),Point(m[0]+(w/2),m[1]+(w/2))).draw(win) # letztes Quadrat wird gemalt
    if w <= 0.1:
        return
    else:
        #Rectangle(Point(m[0]-(w/2),m[1]-(w/2)),Point(m[0]+(w/2),m[1]+(w/2))).draw(win) # "Hauptquadrat" wird gemalt
        drawRectangle((m[0]-(w/2),m[1]-(w/2)),w*0.4) # das linke untere Eck wird als Mittelpunkt genommen
        drawRectangle((m[0]-(w/2),m[1]+(w/2)),w*0.4) # das linke obere Eck wird als Mittelpunkt genommen
        drawRectangle((m[0]+(w/2),m[1]-(w/2)),w*0.4) # das rechte untere Eck wird als Mittelpunkt genommen
        drawRectangle((m[0]+(w/2),m[1]+(w/2)),w*0.4) # das rechte obere Eck wird als Mittelpunkt genommen

drawRectangle((5,5),4)
```

```python
win.close()
```

# 2.4

```python
import matplotlib.pyplot as plt
%matplotlib inline

def line(s:tuple,e:tuple): 
    """ zeichne Linie von Start s bis Ende e"""
    plt.plot([s[0],e[0]],[s[1],e[1]], "b-")

def baum(x:int,y:int,b:int,h:int):
    line((x,y),(x,y+h))
    line((x,y),(x+(b/2),y+h))
    if h <= 1:
        return
    else:
        baum(x,y+h,b/2,h-1)
        baum(x+b/2,y+h,b/2,h-1)

baum(4,1,16,4)
```

# 2.5

```python
from graphics import *
from math import sqrt

win = GraphWin(width = 500, height = 500) # create a window
win.setCoords(0, 0, 10, 10) # set the coordinates of the window; bottom left is (0, 0) and top right is (10, 10)

def sierpinski(x:int, y:int, w:int):
    h = (sqrt(3)/2)*w
    if w <= 0.1:
        l,r,t = Point(x-(w/2),y-(h/3)),Point(x+(w/2),y-(h/3)),Point(x,y+((2*h)/3))
        p = Polygon(l,r,t) # letztes Quadrat wird gemalt
        p.setFill("black")
        p.draw(win)
    else:
        sierpinski(x-(w/4),y-(h/6),w/2)
        sierpinski(x+(w/4),y-(h/6),w/2)
        sierpinski(x,y+(h/3),w/2)

sierpinski(5,5,9)
```

```python
win.close()
```
