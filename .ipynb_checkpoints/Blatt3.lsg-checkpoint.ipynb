# 1

```python
def ins2(l,x):
    for i in range(len(l)):
        if l[i] > x:
            l = l[:i] + [x] + l[i:]
            break
    if x >= l[-1]:
        l = l + [x]
    return l
```

```python
def insSort_R(l): 
    if len(l)<=1: return l
    else: 
        return ins2(insSort_R(l[:-1]), l[-1]) 
```

```python
def insSort_Iter(l):
    erg = [l[0]]
    l = l[1:]
    for u in l:
        for s in range(len(erg)):
            if u < erg[s]:
                erg = erg[:s] + [u] + erg[s:]
                break
        if u >= erg[-1]:
            erg = erg + [u]
    return erg
```

```python
insSort_Iter([3,6,3,1,6])
```

# 2

```python
def insertionSortInplace(l:list) -> list:
    s = [l[0]]
    o = l[1:]
    for e in range(len(o)):
        s = ins2(s,o.pop(0))
    return s
```

```python
insertionSortInplace([3,6,3,1,6])
```

# 3

```python
def quickSort(lst):
    if len(lst) <= 1:
        return lst
    p = lst[0]
    l = list()
    r = list()
    for i in lst[1:]:
        if i < p:
            l.append(i)
        else: 
            r.append(i)
    return quickSort(l) + [p] + quickSort(r)

from numpy.random import choice
import numpy as np

def quickSortRandom(lst,l,r):
    if len(lst) <= 1:
        return lst
    lst = list(lst)
    rand = choice(np.arange(l,r+1))
    p = lst[rand]
    left = list()
    right = list()
    for i in lst[:rand] + lst[rand+1:]:
        if i < p:
            left.append(i)
        else: 
            right.append(i)
    return quickSort(left) + [p] + quickSort(right)

def quickSortMedian(lst,l,r):
    if len(lst) <= 1:
        return lst
    lst = list(lst)
    rand = choice(np.arange(l,r+1),3,replace=False)
    for e in rand: 
        if e != max(rand) and e != min(rand): medi = e
    p = lst[medi]
    left = list()
    right = list()
    for i in lst[:medi] + lst[medi+1:]:
        if i < p:
            left.append(i)
        else: 
            right.append(i)
    return quickSort(left) + [p] + quickSort(right)
```

```python
randomList = np.random.choice(np.arange(1,100001),10000*100).reshape(100,10000)
```

```python
from time import time_ns

quick = list()

for i in randomList:
    start = time_ns()
    quickSort(i)
    end = time_ns()
    quick.append(end-start)

quickRand = list()

for i in randomList:
    start = time_ns()
    quickSortRandom(i,4000,6000)
    end = time_ns()
    quickRand.append(end-start)

quickMed = list()

for i in randomList:
    start = time_ns()
    quickSortMedian(i,4000,6000)
    end = time_ns()
    quickMed.append(end-start)
```

```python
print(np.array(quick).mean())
print(np.array(quickRand).mean())
print(np.array(quickMed).mean())
```

```python
print(min(quick))
print(min(quickRand))
print(min(quickMed))
```

```python
print(max(quick))
print(max(quickRand))
print(max(quickMed))
```
