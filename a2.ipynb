# Aufgabe 1

```python
def sum(n):
    if n == 1:
        return 1
    return n + sum(n-1)

sum(5)
```

```python
def own_len(lst):
    if len(lst) == 1:
        return 1
    return 1 + own_len(lst[:-1])

own_len([1,2,3,4])
```

# Aufgabe 2

```python
import matplotlib.pyplot as plt
from numpy import arange

def line(x,y):
    plt.plot([x,x],[0,y],color="blue") # Linie wird mit bestimmter Höhe an Position X gezeichnet

def lineal_iter(x,y,h):
    for i in arange(0,h): # Schritte der Höhen für die Linien
        for j in arange(x + ((y-x)/(2**h)),y,(y-x)/(2**h)): # Man benötigt immer (2**h) Linien. (y-x)/(2**h) ist dann die Entfernung zwischen den Linien. Die erste Linie hat diese Entfernung auch zu 0 und die letzte zu 100 (bzw zu x und y).
            if (j-x) % ((y-x)/(2**(i+1))) == 0: # Die einzelnen Abstände werden durchlaufen und es wird überprüft, welche Höhe die Linie haben muss. (y-x) / 2**(0+1) wäre also die Hälfte und (y-x) / (2**(1+1)) jeweils die Vierteil etc
                line(j,h-i) # Wenn die Bedingung stimmt, wird dort eine Linie mit entsprechender Höhe gezeichnet.

lineal_iter(0,100,5)
```

# Aufgabe 3

```python
from graphics import *

win = GraphWin(width = 500, height = 500) # create a window
win.setCoords(0, 0, 10, 10) # set the coordinates of the window; bottom left is (0, 0) and top right is (10, 10)
def drawRectangle(m:tuple, w:int):
    Rectangle(Point(m[0]-(w/2),m[1]-(w/2)),Point(m[0]+(w/2),m[1]+(w/2))).draw(win) # Es wird immer ein Rechteck um den übergebenen Mittelpunkt mit ensprechender Breite gezeichnet.
    if w <= 0.1:
        return # Annrichkriterium ist erreicht und die Funktion beendet sich.
    else:
        drawRectangle((m[0]-(w/2),m[1]-(w/2)),w*0.4) # das linke untere Eck wird als Mittelpunkt genommen und 2/5 tel der Ursprungsbreite
        drawRectangle((m[0]-(w/2),m[1]+(w/2)),w*0.4) # das linke obere Eck wird als Mittelpunkt genommen und 2/5 tel der Ursprungsbreite
        drawRectangle((m[0]+(w/2),m[1]-(w/2)),w*0.4) # das rechte untere Eck wird als Mittelpunkt genommen und 2/5 tel der Ursprungsbreite
        drawRectangle((m[0]+(w/2),m[1]+(w/2)),w*0.4) # das rechte obere Eck wird als Mittelpunkt genommen und 2/5 tel der Ursprungsbreite

drawRectangle((5,5),4)
```

```python
win.close()
```

# Aufgabe 4

```python
def line(s:tuple,e:tuple): 
    """ zeichne Linie von Start s bis Ende e"""
    plt.plot([s[0],e[0]],[s[1],e[1]], "b-")
 
def baum(x:int,y:int,b:int,h:int):
    line((x,y),(x,y+h)) # Senkrechter Strich vom Punkt aus h nach oben
    line((x,y),(x+(b/2),y+h)) # Schräger Strich noch um die halbe Breite nach rechts
    if h <= 1:
        return
    else:
        baum(x,y+h,b/2,h-1) # Neuer Unterbaum mit der halben Breite und h-1 vom einen Ende der Striche
        baum(x+b/2,y+h,b/2,h-1) # und vom anderen Ende der Striche

baum(0,0,16,4)
```

# Aufgabe 5

```python
from graphics import *
from math import sqrt

win = GraphWin(width = 500, height = 500) # create a window
win.setCoords(0, 0, 10, 10) # set the coordinates of the window; bottom left is (0, 0) and top right is (10, 10)

def sierpinski(x:int, y:int, w:int):
    h = (sqrt(3)/2) * w # Höhe eines gleichseitigen Dreiecks
    if w <= 0.1: # Abbruchkriterium ist erreicht
        l,r,t = Point(x-(w/2),y-(h/3)),Point(x+(w/2),y-(h/3)),Point(x,y+((2*h)/3)) # Die Eckpunkte werden bestimmt
        p = Polygon(l,r,t) # letztes Dreieck wird gezeichnet (also nur im letzten Schritt wird gezeichnet)
        p.setFill("black")
        p.draw(win)
    else:
        sierpinski(x-(w/4),y-(h/6),w/2) # Aufruf mit den neuen Mittelpunkten und der halben Breite
        sierpinski(x+(w/4),y-(h/6),w/2)
        sierpinski(x,y+(h/3),w/2)

sierpinski(5,5,9)
```

```python
win.close()
```
